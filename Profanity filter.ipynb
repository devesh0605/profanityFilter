```
# importing library
from better_profanity import profanity
```

```
# replacing cuss words with default symbol
text = "you  piece of shit "
censored_text = profanity.censor(text)
print(censored_text)
```

```
# replacing cuss words with custom symbol
text = "dumb cunt"
censored_text = profanity.censor(text,"$")
print(censored_text)
```

```
# checking for cuss words
text = "you  dickhead"
censored_text = profanity.contains_profanity(text)
print(censored_text)
```

```
# adding our own cuss words
text="Let's go to the mall"
result=profanity.censor(text)
print(result)
custom_words=['mall']
profanity.add_censor_words(custom_words)
text="Let's go to the mall"
result=profanity.censor(text)
print(result)
```
